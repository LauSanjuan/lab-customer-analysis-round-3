# Lab | Customer Analysis Round 3. EDA

```python
import pandas as pd
marketing_customer_analysis = pd.read_csv('./files_for_lab/csv_files/marketing_customer_analysis.csv')
marketing_customer_analysis
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# 1. Show DataFrame info.
marketing_customer_analysis.info()
```

```python
# 2. Describe DataFrame.
marketing_customer_analysis.describe()
```

```python
# 3.Show a plot of the total number of responses.
fig, ax = plt.subplots()
ax.bar(x = marketing_customer_analysis['Response'].value_counts().index, height = marketing_customer_analysis['Response'].value_counts().values, color = ['steelblue', 'grey'])
ax.set_xlabel("Response")
ax.set_ylabel("Number of responses")
ax.set_title("Total number of responses")
plt.show()
```

```python
# 4. Show a plot of the response rate by the sales channel.
_ = marketing_customer_analysis.groupby(['Sales Channel','Response']).size().unstack().plot(kind='bar', title='Response rate by sales channel',ylabel='Number of responses',stacked=True)
```

```python
# 5. Show a plot of the response rate by the total claim amount.
marketing_customer_analysis['Range'] = pd.qcut(marketing_customer_analysis['Total Claim Amount'], q=5)
_ = marketing_customer_analysis.groupby(['Range','Response']).size().unstack().plot(kind='bar', title='Response rate by total claim amount',xlabel='Total Claim Amount',ylabel='Number of responses',stacked=True)
```

```python
# 6. Show a plot of the response rate by income.
marketing_customer_analysis['Income Range'] = pd.qcut(marketing_customer_analysis['Income'], q=3)
_ = marketing_customer_analysis.groupby(['Income Range','Response']).size().unstack().plot(kind='bar', title='Response rate by income',xlabel='Income',ylabel='Number of responses',stacked=True)
```
